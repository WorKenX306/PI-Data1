```python
import pandas as pd
from pathlib import Path

PROJECT_ROOT = Path(__file__).parent.parent if "__file__" in globals() else Path("..")
reports_dir = PROJECT_ROOT / "reports" / "mlops"

metrics_df = pd.read_csv(reports_dir / "metrics_latest.csv")
best_df = pd.read_json(reports_dir / "best_models_latest.json")
checks_df = pd.read_csv(reports_dir / "artifact_checks_latest.csv")
```

```python
print("All Model Metrics:")
display(metrics_df)

print("\nBest Models per Dataset:")
display(best_df)
```

```python
print("Artifact Check Results:")
display(checks_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# F1 score comparison per model
sns.barplot(data=metrics_df, x="model_name", y="f1_macro", hue="dataset")
plt.xticks(rotation=45)
plt.title("F1 Macro Score by Model & Dataset")
plt.show()
```
